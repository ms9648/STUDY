```python
import pandas as pd
import numpy as np

movie = pd.read_csv("../ml-latest-small/movies.csv")
movie = movie.set_index("movieId")
movie['year'] = movie['title'].str.extract("(\(\d\d\d\d\))")
movie['year'] = movie['year'].str.extract("(\d\d\d\d)")
movie.sample(5, random_state = 42)
movie['year'] = movie['year'].fillna(np.nan)

genres_unique = []
for item in list(movie['genres'].apply(lambda x: x.split('|'))):
    for i in item:
        genres_unique.append(i)

genres_unique = list(set(genres_unique))

genres = movie['genres'].str.get_dummies(sep = "|")
```

```python
ratings = pd.read_csv("../ml-latest-small/ratings.csv")
ratings = ratings.set_index('userId')

ratings.sample(5, random_state=42)
```

```python

UserProfile = ratings.merge(genres, left_on='movieId', right_index=True)
UserProfile = UserProfile.reset_index()
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(UserProfile, random_state = 42, test_size = .1)

for col in genres.columns:
    train[col] = train[col] * train['rating']

UserProfile = train.groupby('userId')[genres.columns].mean()
UserProfile
```
